```python
x <-1:12600 #just a list of numbers, each corresponding to a gene

labels <- df[,12601] #extracting labels as dataframe (just in case)
entropyvec <- vector(mode = "numeric", length = 12600) #to store entropies of each column
varvac <- vector(mode = "numeric", length = 12600) #to store variance of each column
```

```python
# index_usefulgenes <- 1
for (val in x) {
    a <- max(df[,c(val)])
    b <- min(df[,c(val)])
    if(as.integer(a - b) > 0){
        generange <- as.integer(a - b)
    }
    else{
        generange <- 1
    }
     #getting number of bins
    y <- discretize(df[,val], numBins = generange) #discretize function on the variable
    entropyvec[val] <- entropy(y)
    varvac[val] <- var(y)
#     if (entropy(y) > 1.5) {
#         usefulgenes[index_usefulgenes] <- val #if entropy is > 1.5, add the col no to list of useful col nos
#         index_usefulgenes <- index_usefulgenes + 1   #increment index of useful genes vector
#     }
}
```

```python
med <- median(entropyvec)
avg <- mean(entropyvec)
med
avg
```

```python
#visualisation
#SCATTER PLOT

wdata <- data.frame(gene_nums = x, entropies = entropyvec)

b <- ggplot(wdata, aes(gene_nums, entropyvec, color = entropyvec)) +
  geom_point(shape = 16, size = 1, alpha =0.6) +
  theme_minimal() + scale_color_gradient(low = "#0091ff", high = "#f0650e")
b
```

```python

#FREQUENCY PLOT


a <- ggplot(wdata, aes(x = entropyvec))

a + geom_histogram(bins = 25, color = "dark blue", fill = "white") +
  geom_vline(aes(xintercept = mean(entropyvec)), 
             linetype = "dashed", size = 0.6, color = "red") + geom_vline(aes(xintercept = median(entropyvec)), 
             linetype = "dashed", size = 0.6, color = "green") 
```

```python
#doing the actual filtering
usefulgenes <- vector(mode="numeric", length=500) #initializing the vector for the useful genes
index_usefulgenes <- 1
for(val in x){
    if(entropyvec[val] >= 4.75){
        usefulgenes[index_usefulgenes] <- val #if entropy is > MEDIAN, add the col no to list of useful col nos
        index_usefulgenes <- index_usefulgenes + 1   #increment index of useful genes vector
    }
    
}
```
